## Introduction to NumPy


### What is NumPy?

NumPy stands for Numerical Python. It is a powerful Python library used for numerical and mathematical operations. It is mainly used for working with arrays and performing fast calculations.

NumPy was created in 2005 by Travis Oliphant. It is an open source project and you can use it freely.

The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.
NumPy is widely used in:

- Data Science
- Machine Learning
- Artificial Intelligence
- Scientific Computing
- Data Analysis
- image processing
- stock market
- Medical Research

### Why Do We Need NumPy?

Python has built-in lists. So why use NumPy?

Because NumPy arrays are:

1. **Faster** – NumPy is written in C language internally.
2. **Memory Efficient** – It uses less memory compared to Python lists.
3. **Better for Math Operations** – Supports vector and matrix operations.
4. **Multi-dimensional** – Supports 1D, 2D, and n-dimensional arrays.

### Import NumPy

Once NumPy is installed, import it in your applications by adding the import keyword:

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print(arr)
```

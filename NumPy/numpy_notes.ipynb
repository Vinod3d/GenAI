### Indexing
Indexing in NumPy means accessing specific elements from an array using their position (index). It is very similar to Python lists but more powerful.

#### Indexing in 1D Array

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr[0])   # First element → 10
print(arr[2])   # Third element → 30
print(arr[-1])  # Last element → 50
```

#### Indexing in 2D Array


```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print(arr[0, 0])  # Row 0, Column 0 → 1
print(arr[1, 2])  # Row 1, Column 2 → 6
```

#### Indexing in 3D Array

```python
arr = np.array([
    [[1, 2, 3],
     [4, 5, 6]]
])

print(arr[0, 1, 2])  # → 6
```

#### Fancy Indexing

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

result = arr[[0, 2, 4]]

print(result)   # [10 30 50]
```

### Slicing
Like lists, we can slice the numpy array. Slicing in python means taking the elements from the given index to another given index.

We pass slice like this: [start:end].The element at end index is not being included in the output.

We can select the elements from 1 to 3 and assign it to a new numpy array d as follows:

```python
# Slicing the numpy array

d = c[1:4]
d
```

```python
# Set the fourth element and fifth element to 300 and 400

c[3:5] = 300, 400
c
```

We can also define the steps in slicing, like this: [start:end:step].

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[1:5:2])
```

If we don't pass start its considered 0

```python
print(arr[:4])
```

If we don't pass end it considers till the length of array.

```python
print(arr[4:])
```

If we don't pass step its considered 1

```python
print(arr[1:5:])
```

#### Aggregation function 
Aggregation functions in NumPy are used to perform operations on multiple values and return a single summarized result. These are very useful in data analysis when you want to calculate totals, averages, minimums, etc.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7,])

print("Sum:", np.sum(arr))         # 15
print("Mean:", np.mean(arr))       # 3.0
print("Min:", np.min(arr))         # 1
print("Max:", np.max(arr))         # 5
print("Std:", np.std(arr))         # Standard deviation
print("variance", np.var(arr)) # Variance tells us how spread out the values are from the average (mean).
print("Product", np.prod(arr)) # It multiplies all values in the array and returns a single result.
print("cumulative_sum", np.cumsum(arr)) #It keeps adding elements one by one and stores the result at each step.
print("cumulative_product", np.cumprod(arr)) # Instead of multiplying all numbers at once, it multiplies step by step and stores each result.
```

### Reshaping in NumPy
Reshaping means changing the shape (dimensions) of an array without changing its data.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

reshaped = arr.reshape(2, 3)

print(reshaped)
```

```python
# 1D → 2D
arr = np.array([1, 2, 3, 4])
print(arr.reshape(2, 2))
```

```python
# 2D → 1D (Flatten)

arr = np.array([[1, 2], [3, 4]])
print(arr.reshape(-1))   # [1 2 3 4]
```

```python
# Reshaping into 3D
arr = np.array([1, 2, 3, 4, 5, 6])

print(arr.reshape(1, 2, 3))
```

```python
# .ravel() Changes in r may affect original array
import numpy as np
arr = np.array([[1, 2], [3, 4]])
r = arr.ravel()
print(r)   # [1 2 3 4]
```

```python
# .flatten() Changes in f do NOT affect original array
arr = np.array([[1, 2], [3, 4]])
f = arr.flatten()
print(f)   # [1 2 3 4]
```

### Insert
It allows you to insert values at specific positions in 1D, 2D, or higher-dimensional arrays.

```python
# Insert in 1D Array

import numpy as np
arr = np.array([1, 2, 3, 4])
new_arr = np.insert(arr, 2, 99)
print(new_arr)   # [1 2 99 3 4]
```

```python
# Insert Multiple Values

arr = np.array([1, 2, 3])
new_arr = np.insert(arr, 1, [10, 20])
print(new_arr)   # [1 10 20 2 3]
```

```python
# Insert in 2D Array (Row-wise)
arr = np.array([[1, 2],
                [3, 4]])
new_arr = np.insert(arr, 1, [9, 9], axis=0)
print(new_arr)
```

### Append

### Concatenate
Concatenate means joining two or more arrays together.

```python
# Concatenate in 1D Array

import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = np.concatenate((arr1, arr2))
print(result)   # [1 2 3 4 5 6]
```

```python
# Concatenate in 2D Array (Row-wise)

arr1 = np.array([[1, 2],
                 [3, 4]])
arr2 = np.array([[5, 6]])
result = np.concatenate((arr1, arr2), axis=0)
print(result)
```

### Stacking
Stacking means combining multiple arrays along a new axis (dimension).
- np.stack()	General stacking (new axis)
- np.vstack()	Vertical stacking (row-wise)
- np.hstack()	Horizontal stacking (column-wise)

```python
# np.stack()

import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = np.stack((arr1, arr2))

print(result)
```

```python
# np.vstack()

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = np.vstack((arr1, arr2))

print(result)
```

```python
# np.hstack()

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = np.hstack((arr1, arr2))

print(result)
```

### Broadcasting

Broadcasting is a powerful feature in NumPy that allows arrays of different shapes to perform operations together without explicitly reshaping them.

```python
import numpy as np
arr = np.array([1, 2, 3])
result = arr + 5

print(result)   # [6 7 8]
```

👉 Here, 5 is automatically applied to all elements

👉 NumPy "broadcasts" the scalar to match array size

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([10, 20, 30])

print(arr1 + arr2)   # [11 22 33]
```

```python
import numpy as np
prices = np.array([100, 200, 300])
discount = 30
final_prices = prices - (prices * discount/100)

print(final_prices)
```

```python
arr1 = np.array([[1, 2, 3],
                 [4, 5, 6]])
arr2 = np.array([[10],
                 [20]])

print(arr1 + arr2)
```

### Vectorization vs Broadcasting

### Vectorization
Vectorization means performing operations on entire arrays at once without using explicit loops.

It improves performance and speed

Uses optimized NumPy internal implementation (C-level)

### Broadcasting

Broadcasting is a technique that allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array.

Makes arrays shape-compatible

Happens before vectorized operation


### Handling Missing Values
In NumPy, missing values are usually represented as:

np.nan → Not a Number (for float arrays)

#### Creating Missing Values

```python
import numpy as np

arr = np.array([1, 2, np.nan, 4, np.nan])

print(arr)
# [ 1.  2. nan  4. nan]
```

#### Detect Missing Values

```python

print(np.isnan(arr))
# [False False True False True]
```

#### Replace Missing Values with 0


```python
cleaned_arr = np.nan_to_num(arr, nan=100)
cleaned_arr
```

#### Replace Inf


```python
import numpy as np

arr = np.array([1, 2, np.inf, 4, -np.inf, 6])

print(np.isinf(arr))

cleaned_arr = np.nan_to_num(arr, posinf=1000, neginf=-1000)

print(cleaned_arr)
```

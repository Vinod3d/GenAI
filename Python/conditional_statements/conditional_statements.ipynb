## 📖 Introduction

### What are Conditional Statements?
Conditional statements are blocks of code that run only when a specific condition is met. Think of them as "decision points" in your program.

### Why do we use them?
In real life, we make decisions based on conditions:
* **If** the alarm rings, **then** we wake up.
* **If** it is raining, **then** we carry an umbrella.

In programming, we need our software to be smart and react dynamically. For example, a website only logs you in **if** your password is correct. Conditional statements allow our programs to choose different paths based on different inputs.


## 🔹 The `if` Statement

### Explanation
The `if` statement is the most basic decision-making statement. It executes a block of code only if a specified condition evaluates to `True`. If the condition is `False`, the code block is skipped.

### Syntax
```python
if condition:
    # Code to run if the condition is True
```


### 💡 Simple Example: The `if` Statement


````python
# Define a number variable
number = 15

# Check if the number is greater than 10
if number > 10:
    # This block runs because 15 is greater than 10
    print("The number is greater than 10.")
````

**Expected Output:**
```text
The number is greater than 10.
```


### 🌍 Real-Life Example: The `if` Statement


````python
# Check if a user is old enough to vote
user_age = 20

# The legal voting age is 18
if user_age >= 18:
    # This block runs because 20 is greater than or equal to 18
    print("You are eligible to register for voting!")
````

**Expected Output:**
```text
You are eligible to register for voting!
```

## 🔹 The `if-else` Statement

### Explanation
The `if-else` statement provides an alternative path. If the condition is `True`, the code block under `if` is run. If the condition is `False`, the code block under `else` runs instead.

### Syntax
```python
if condition:
    # Code runs if condition is True
else:
    # Code runs if condition is False
```


### 💡 Simple Example: The `if-else` Statement


````python
# Define a number variable
number = 7

# Check if the remainder when dividing by 2 is 0
if number % 2 == 0:
    print("The number is even.")
else:
    # This block runs because 7 is not divisible by 2
    print("The number is odd.")
````

**Expected Output:**
```text
The number is odd.
```


### 🌍 Real-Life Example: The `if-else` Statement


````python
# Check if a customer has enough balance to buy an item
account_balance = 120.50
item_price = 150.00

# Compare balance with the price of the item
if account_balance >= item_price:
    print("Purchase successful! Thank you for your purchase.")
else:
    # This block runs because 120.50 is less than 150.00
    print("Declined: Insufficient funds in your account.")
````

**Expected Output:**
```text
Declined: Insufficient funds in your account.
```

## 🔹 The `if-elif-else` Statement

### Explanation
The `if-elif-else` statement lets you check multiple conditions in sequence. Python checks them from top to bottom. The first condition that is `True` will have its code block executed, and all remaining conditions will be skipped.

### Syntax
```python
if condition1:
    # Runs if condition1 is True
elif condition2:
    # Runs if condition1 is False and condition2 is True
else:
    # Runs if all above conditions are False
```


### 💡 Simple Example: The `if-elif-else` Statement


````python
# Compare a number to zero
number = -5

# Check positive, negative, or zero
if number > 0:
    print("The number is positive.")
elif number < 0:
    # This block runs because -5 is less than 0
    print("The number is negative.")
else:
    print("The number is zero.")
````

**Expected Output:**
```text
The number is negative.
```


### 🌍 Real-Life Example: The `if-elif-else` Statement


````python
# Calculate movie ticket price based on age group
visitor_age = 15

# Determine price based on age boundaries
if visitor_age < 5:
    print("Ticket Price: Free")
elif visitor_age <= 18:
    # This block runs because 15 is between 5 and 18
    print("Ticket Price: $10 (Child/Teen Ticket)")
elif visitor_age < 60:
    print("Ticket Price: $20 (Adult Ticket)")
else:
    print("Ticket Price: $12 (Senior Ticket)")
````

**Expected Output:**
```text
Ticket Price: $10 (Child/Teen Ticket)
```

## 🔹 Nested `if` Statements

### Explanation
A nested `if` is an `if` statement placed inside another `if` statement. This is useful when a second condition needs to be checked only after the first condition has successfully passed.

### Syntax
```python
if outer_condition:
    # Runs if outer_condition is True
    if inner_condition:
        # Runs if BOTH outer and inner conditions are True
```


### 💡 Simple Example: Nested `if` Statements


````python
# Check if a number is positive, and if so, check if it is even
number = 14

# Outer check: Is the number positive?
if number > 0:
    print("The number is positive.")
    
    # Inner check: Is it also even?
    if number % 2 == 0:
        # This block runs because 14 is positive AND even
        print("It is also an even number.")
````

**Expected Output:**
```text
The number is positive.
It is also an even number.
```


### 🌍 Real-Life Example: Nested `if` Statements


````python
# E-commerce store checkout login and membership status check
is_logged_in = True
is_premium_member = True

# Outer check: User must be logged in
if is_logged_in:
    print("Welcome back to the online store!")
    
    # Inner check: Does the user have premium membership?
    if is_premium_member:
        # This block runs because the user is logged in AND is premium
        print("Premium discount applied: 15% off your order.")
````

**Expected Output:**
```text
Welcome back to the online store!
Premium discount applied: 15% off your order.
```

## 🔹 Short-hand `if` Statement

### Explanation
If you have only a single line of code to execute inside an `if` statement, you can write the condition and the statement on the same line to keep the code compact.

### Syntax
```python
if condition: statement
```


### 💡 Simple Example: Short-hand `if` Statement


````python
# Short-hand check if score qualifies for top grade
student_score = 95

# Print output on the same line as the condition
if student_score >= 90: print("Grade: A")
````

**Expected Output:**
```text
Grade: A
```


### 🌍 Real-Life Example: Short-hand `if` Statement


````python
# Check if a system battery is critically low
battery_level = 12

# Alert immediately on a single line if battery is below 15%
if battery_level < 15: print("Battery low alert: Please connect your charger.")
````

**Expected Output:**
```text
Battery low alert: Please connect your charger.
```

## 🔹 Short-hand `if-else` (Ternary Operator)

### Explanation
The short-hand `if-else` (known as the ternary operator) allows you to write a simple `if-else` choice on a single line. It is commonly used to assign a value to a variable based on a condition.

### Syntax
```python
value_if_true if condition else value_if_false
```


### 💡 Simple Example: Short-hand `if-else` (Ternary Operator)


````python
# Classify a number as positive or negative on a single line
number = -8

# Ternary assignment
result = "Positive" if number >= 0 else "Negative"
print("The number is:", result)
````

**Expected Output:**
```text
The number is: Negative
```


### 🌍 Real-Life Example: Short-hand `if-else` (Ternary Operator)


````python
# Check order total for free shipping qualification
total_order_amount = 75.00

# Qualifies for free shipping if spending is $50.00 or more
shipping_type = "Free Shipping" if total_order_amount >= 50.00 else "Standard Shipping ($5.99)"
print("Shipping Method:", shipping_type)
````

**Expected Output:**
```text
Shipping Method: Free Shipping
```

## 🔹 Logical Operators (`and`, `or`, `not`) in Conditions

### Explanation
Logical operators allow you to combine or modify multiple conditions. Use `and` when all conditions must be true, `or` when at least one condition must be true, and `not` to flip or invert a condition's truth value.

### Syntax
```python
# Using 'and'
if condition1 and condition2:
    # Runs if both are True

# Using 'or'
if condition1 or condition2:
    # Runs if at least one is True

# Using 'not'
if not condition:
    # Runs if condition is False
```


### 💡 Simple Example: Logical Operators (`and`, `or`, `not`) in Conditions


````python
# Check if a number is inside the range of 10 to 20 inclusive
value = 15

# Both parts of the 'and' must be True
if value >= 10 and value <= 20:
    print("The value is within the range [10, 20].")
````

**Expected Output:**
```text
The value is within the range [10, 20].
```


### 🌍 Real-Life Example: Logical Operators (`and`, `or`, `not`) in Conditions


````python
# Check database access for user: must have valid password OR admin clearance
has_valid_password = False
is_administrator = True

# Access is granted if either condition is True
if has_valid_password or is_administrator:
    print("Access granted: Welcome to the administration dashboard.")
else:
    print("Access denied.")
````

**Expected Output:**
```text
Access granted: Welcome to the administration dashboard.
```

## 🔹 Comparison Operators in Conditions

### Explanation
Comparison operators compare two values and return `True` or `False`. They form the core of conditional checks. The main comparison operators are: `==` (equal to), `!=` (not equal to), `>` (greater than), `<` (less than), `>=` (greater than or equal to), and `<=` (less than or equal to).

### Syntax
```python
x == y  # Equal to
x != y  # Not equal to
x > y   # Greater than
x < y   # Less than
x >= y  # Greater than or equal to
x <= y  # Less than or equal to
```


### 💡 Simple Example: Comparison Operators in Conditions


````python
# Compare two variables for inequality
first_value = 100
second_value = 200

# Check if the values are different
if first_value != second_value:
    print("first_value and second_value are different.")
````

**Expected Output:**
```text
first_value and second_value are different.
```


### 🌍 Real-Life Example: Comparison Operators in Conditions


````python
# Check if room temperature is exactly at set target
target_temp = 22
current_temp = 22

# Checks if current temperature equals the target temperature
if current_temp == target_temp:
    print("Target temperature met. Thermostat on standby.")
````

**Expected Output:**
```text
Target temperature met. Thermostat on standby.
```

```python
# 1. Using type() to check the exact type of a variable
x = 42
y = [1, 2, 3]
print(f"type(x): {type(x)}")
print(f"type(y): {type(y)}")

# 2. Using isinstance() - Recommended for validation (supports inheritance checking)
class Animal: pass
class Dog(Animal): pass

my_dog = Dog()
print(f"Is my_dog an instance of Dog? {isinstance(my_dog, Dog)}")
print(f"Is my_dog an instance of Animal? {isinstance(my_dog, Animal)}") # True due to inheritance
print(f"Is type(my_dog) exactly Animal? {type(my_dog) is Animal}")        # False
```

## 3️⃣ Mutable vs. Immutable Objects

One of the most frequent Python interview topics is the difference between mutable and immutable data types.

### 💡 The Core Difference:
* **Immutable Objects**: Once created, their value **cannot** be changed. Modifying them creates a *new object* in memory (with a different `id()`).
  * *Examples*: `int`, `float`, `complex`, `str`, `tuple`, `bool`, `frozenset`, `bytes`.
* **Mutable Objects**: Their value **can** be changed in-place without changing their memory address (`id()`).
  * *Examples*: `list`, `dict`, `set`, `bytearray`.

### 🔑 Why this distinction matters in interviews:
1. **Dictionary Keys & Sets**: Dictionary keys and set items must be **hashable** (which implies they must be immutable). Using a mutable object like a `list` as a dictionary key raises a `TypeError`.
2. **Thread Safety**: Immutable objects are inherently thread-safe because they cannot be modified after creation.
3. **Function Arguments**: Passing mutable objects to functions can lead to side-effects because modifications inside the function affect the caller's object.

```python
# --- IMMUTABLE TYPE (Integer) DEMO ---
a = 10
print(f"Initial a: {a}, Address (id): {id(a)}")
a = a + 1
# Notice that the memory address changes because integers are immutable!
print(f"Updated a: {a}, Address (id): {id(a)}")

print("-" * 40)

# --- MUTABLE TYPE (List) DEMO ---
my_list = [1, 2, 3]
print(f"Initial list: {my_list}, Address (id): {id(my_list)}")
my_list.append(4)
# Memory address remains exactly the same! The object was modified in-place.
print(f"Updated list: {my_list}, Address (id): {id(my_list)}")
```

## 4️⃣ Sequence Types & Memory Optimizations

### ⚖️ List vs. Tuple:
| Feature | List (`list`) | Tuple (`tuple`) |
| :--- | :--- | :--- |
| **Mutability** | **Mutable** (can append, insert, remove). | **Immutable** (fixed size and elements). |
| **Memory** | Larger (uses over-allocation to make appends faster). | Smaller (fixed size allocated exactly). |
| **Performance**| Slower creation and iteration. | Faster creation and iteration. |
| **Use Case** | Heterogeneous/Homogeneous data expected to change. | Read-only structures, coordinates, database rows. |

### 🧠 Under the Hood: Caching & Interning
To optimize performance, Python employs:
* **Integer Caching**: Python pre-allocates and caches integers in the range `[-5, 256]`. Any variables pointing to these numbers will reference the *same* object in memory.
* **String Interning**: Short strings (matching variable name patterns) are interned automatically. You can explicitly intern strings using `sys.intern()`.

```python
import sys

# 1. Memory footprint comparison: List vs Tuple
temp_list = [1, 2, 3, 4, 5]
temp_tuple = (1, 2, 3, 4, 5)
print(f"Memory size of List: {sys.getsizeof(temp_list)} bytes")
print(f"Memory size of Tuple: {sys.getsizeof(temp_tuple)} bytes")

print("-" * 40)

# 2. Integer Caching Demo (Range [-5, 256])
x = 256
y = 256
print(f"x is y (256): {x is y}") # True: cached

a = 257
b = 257
print(f"a is b (257): {a is b}") # False: not cached (different memory addresses)
```

## 5️⃣ Unordered Collections: Dictionaries & Sets

Both `dict` and `set` rely on **hash tables** for their implementation, offering highly optimized lookup times.

### 🔑 Key Characteristics:
* **Time Complexity**: Average time complexity for insertion, deletion, and search is **$O(1)$** (constant time). In contrast, list search is **$O(n)$** (linear time).
* **Hashability Requirement**: For an object to be hashed, its hash value must never change during its lifetime. Hence, only immutable objects can be dict keys or set items.
* **Ordering**:
  * **Set**: Unordered. Order of elements is not guaranteed.
  * **Dictionary**: Since Python 3.7, dictionaries maintain the **insertion order** as an implementation detail, but they are still optimized mapping structures.

```python
# 1. Standard Set operations (Union, Intersection, Difference)
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print(f"Union (set_a | set_b): {set_a | set_b}")
print(f"Intersection (set_a & set_b): {set_a & set_b}")
print(f"Difference (set_a - set_b): {set_a - set_b}")

print("-" * 40)

# 2. Demonstrating the Hashability requirement (uncommenting raises TypeError)
try:
    bad_dict = {
        [1, 2]: "Invalid key!" # List is mutable, so it is unhashable
    }
except TypeError as e:
    print(f"Caught expected error: {e}")
```

## 6️⃣ Must-Know Interview Questions on Data Types

### Q1: What is the difference between `is` and `==`?
* **`==` (Equality Operator)**: Compares the **values** of two objects to see if they are equal.
* **`is` (Identity Operator)**: Compares the **memory locations (addresses)** of two objects. It returns `True` only if both variables point to the exact same object in memory.

---

### Q2: Explain the "Mutable Default Argument Trap" in functions.
In Python, default arguments are evaluated **once** when the function is defined, not each time the function is called.
If you use a mutable object (like a list) as a default argument, that single list is shared across all function calls that don't provide an argument.

---

### Q3: What is the difference between Shallow Copy and Deep Copy?
* **Shallow Copy (`copy.copy()`)**: Creates a new container object, but fills it with references to the nested objects from the original. Modifying nested structures affects both.
* **Deep Copy (`copy.deepcopy()`)**: Recursively copies the container *and* all objects nested inside it. The new object is entirely independent of the original.

```python
# 1. 'is' vs '==' demo
list_1 = [1, 2, 3]
list_2 = [1, 2, 3]
print(f"list_1 == list_2: {list_1 == list_2}") # True (values are same)
print(f"list_1 is list_2: {list_1 is list_2}") # False (different objects in memory)

print("-" * 40)

# 2. Mutable Default Argument Trap demo
def append_to_list(val, my_list=[]):
    my_list.append(val)
    return my_list

print("Call 1:", append_to_list(10)) # [10]
print("Call 2:", append_to_list(20)) # [10, 20] (Trap! The list is shared)

# Correct implementation:
def safe_append_to_list(val, my_list=None):
    if my_list is None:
        my_list = []
    my_list.append(val)
    return my_list

print("Safe Call 1:", safe_append_to_list(10)) # [10]
print("Safe Call 2:", safe_append_to_list(20)) # [20] (Fixed!)

print("-" * 40)

# 3. Shallow Copy vs Deep Copy demo
import copy
original = [[1, 2], [3, 4]]

# Shallow copy
shallow = copy.copy(original)
# Deep copy
deep = copy.deepcopy(original)

original[0][0] = 99
print(f"Original after change: {original}")
print(f"Shallow copy (affected): {shallow}") # [0][0] is changed to 99!
print(f"Deep copy (unaffected):  {deep}")    # remains 1!
```

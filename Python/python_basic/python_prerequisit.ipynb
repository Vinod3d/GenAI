## 1️⃣ Setup of a Python Project

A virtual environment is a self-contained directory tree that contains a Python installation for a particular version of Python, plus a number of additional packages. It is crucial to use environments to prevent version conflicts between dependencies of different projects.

### Conda Environment Creation
Using **Conda**, you can create an environment in a specific folder path (e.g., `./venv` inside your project directory) to keep it local to the project.

### Step-by-Step Conda Commands

1. **Navigate to your project directory**:
   ```bash
   cd Python
   ```

2. **Create the environment**:
   Create a conda environment inside a local folder named `venv` with a specific Python version (e.g., 3.10):
   ```bash
   conda create -p venv python=3.10 -y
   ```
   *Note: `-p venv` specifies that the environment files should be created in the `venv` directory of the current folder.*

3. **Activate the environment**:
   ```bash
   # On Windows (Command Prompt / PowerShell)
   conda activate .\venv
   
   # Or using directory prefix / relative path
   conda activate venv/
   ```

4. **Verify the installation**:
   To confirm Python is running from the virtual environment and check its version:
   ```bash
   # Find where the Python executable is located
   where python
   
   # Check the Python version (correct syntax uses double dashes --version)
   python --version
   ```

> 💡 **Tip: Alternative using built-in Python `venv`**
> If you are not using Conda, you can use Python's built-in `venv` module:
> ```bash
> # Create virtual environment
> python -m venv venv
>
> # Activate on Windows
> .\venv\Scripts\activate
> ```

Let's execute your first Python statement. In Python, the `print()` function is used to output text/data to the console.

### 🔑 Key Python Syntax Rules:
1. **No Semicolons**: Python doesn't require a semicolon `;` at the end of statements.
2. **Indentation**: Python uses indentation (spaces or tabs) to define blocks of code (like loops or functions) instead of curly braces `{}`.
3. **Comments**: Single line comments start with `#`. Multi-line strings (docstrings) are enclosed in triple quotes `"""`.

## 3️⃣ requirements.txt file

A `requirements.txt` file is used to list all the external libraries (packages) that your project depends on. This makes it easy for others (and deployment tools) to recreate the exact environment required.

### How to work with `requirements.txt`

1. **Installing Dependencies**:
   To install all packages listed in a `requirements.txt` file, run:
   ```bash
   pip install -r requirements.txt
   ```

2. **Generating the file**:
   To automatically save all currently installed packages in your active environment to a file:
   ```bash
   pip freeze > requirements.txt
   ```

### Example content of a `requirements.txt` file:
```text
numpy==1.24.3
pandas==2.0.2
requests==2.31.0
matplotlib==3.7.1
```

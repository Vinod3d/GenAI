## 📖 Introduction

### What are Loops?
In programming, a **loop** is a control structure that repeats a block of instructions or code multiple times. Instead of writing the same line of code over and over, you can use a loop to do it automatically.

### Why do we use loops?
Imagine you need to print the message `"Hello"` 100 times, or calculate the salary of 500 employees. Writing 100 or 500 lines of identical code is time-consuming and leads to mistakes. Loops solve this by running a code block repeatedly with a single, simple structure.

### Advantages of using loops:
1. **Reduces Code Duplication**: Writes code once and runs it many times.
2. **Saves Time and Effort**: Makes updates easy; changes in one place affect all repetitions.
3. **Improves Readability**: Keeps your files short, clean, and organized.
4. **Dynamic Execution**: Handles collections of varying sizes (e.g., reading files or database rows) automatically.


## 🔹 The `for` Loop

### Explanation
A `for` loop is used to iterate over a sequence (such as a list, string, or range of numbers). It executes a block of code a set number of times, once for each item in the sequence.

### Syntax
```python
for item in sequence:
    # Code to run for each item
```


### 💡 Simple Example: The `for` Loop


````python
# Loop through a list of numbers from 1 to 5
for number in [1, 2, 3, 4, 5]:
    # Print each number
    print("Current number:", number)
````

**Expected Output:**
```text
Current number: 1
Current number: 2
Current number: 3
Current number: 4
Current number: 5
```


### 🌍 Real-Life Example: The `for` Loop


````python
# Send newsletters to a list of email subscribers
subscribers = ["Alice", "Bob", "Charlie"]

for name in subscribers:
    # Print a personalized notification message
    print("Notification sent to " + name + ": Welcome to our newsletter!")
````

**Expected Output:**
```text
Notification sent to Alice: Welcome to our newsletter!
Notification sent to Bob: Welcome to our newsletter!
Notification sent to Charlie: Welcome to our newsletter!
```

## 🔹 The `while` Loop

### Explanation
A `while` loop runs a block of code as long as a specified condition is `True`. It checks the condition before executing the block each time, and terminates as soon as the condition becomes `False`.

### Syntax
```python
while condition:
    # Code runs while condition is True
```


### 💡 Simple Example: The `while` Loop


````python
# Countdown from 3 to 1
count = 3

while count > 0:
    print("Countdown:", count)
    # Decrease count by 1 to eventually end the loop
    count = count - 1
````

**Expected Output:**
```text
Countdown: 3
Countdown: 2
Countdown: 1
```


### 🌍 Real-Life Example: The `while` Loop


````python
# Simulate draining a battery while playing a game
battery_level = 15

# Loop runs while battery is above 5%
while battery_level > 5:
    print("Playing game... Battery level:", battery_level, "%")
    # Battery drains by 4% with each level
    battery_level -= 4

print("Warning: Battery level critical! Game paused.")
````

**Expected Output:**
```text
Playing game... Battery level: 15 %
Playing game... Battery level: 11 %
Playing game... Battery level: 7 %
Warning: Battery level critical! Game paused.
```

## 🔹 Difference Between `for` and `while` Loops

### Explanation
Use a `for` loop when you know the exact number of iterations beforehand (definite iteration). Use a `while` loop when the number of repetitions is unknown and depends on a dynamic condition (indefinite iteration).

### Syntax
```python
# For Loop (Definite iteration)
for item in sequence:
    # Code

# While Loop (Indefinite iteration)
while condition:
    # Code
```


### 💡 Simple Example: Difference Between `for` and `while` Loops


````python
# Compare definite loop (For) vs condition loop (While)

print("--- Running For Loop (prints exactly 3 times) ---")
for i in range(1, 4):
    print("For loop index:", i)

print("\n--- Running While Loop (runs until count exceeds 3) ---")
count = 1
while count <= 3:
    print("While loop count:", count)
    count += 1
````

**Expected Output:**
```text
--- Running For Loop (prints exactly 3 times) ---
For loop index: 1
For loop index: 2
For loop index: 3

--- Running While Loop (runs until count exceeds 3) ---
While loop count: 1
While loop count: 2
While loop count: 3
```


### 🌍 Real-Life Example: Difference Between `for` and `while` Loops


````python
# For loop: Baking a fixed list of 3 pizza orders
print("--- Chef's Task list (For Loop) ---")
pizzas = ["Cheese", "Veggie", "Pepperoni"]
for pizza in pizzas:
    print("Baking a", pizza, "pizza.")

# While loop: Baking pizzas until the chef runs out of dough
print("\n--- Chef's Supply depletion (While Loop) ---")
has_dough = True
pizza_count = 1

while has_dough:
    print("Baking pizza number", pizza_count)
    pizza_count += 1
    if pizza_count > 2:
        # Chef runs out of dough after 2 pizzas
        has_dough = False
        print("Out of dough! Cannot bake more.")
````

**Expected Output:**
```text
--- Chef's Task list (For Loop) ---
Baking a Cheese pizza.
Baking a Veggie pizza.
Baking a Pepperoni pizza.

--- Chef's Supply depletion (While Loop) ---
Baking pizza number 1
Baking pizza number 2
Out of dough! Cannot bake more.
```

## 🔹 The `range()` Function

### Explanation
The `range()` function generates a sequence of numbers. It is highly optimized and commonly used with `for` loops. It takes three parameters: start (inclusive), stop (exclusive), and step (interval).

### Syntax
```python
range(stop)             # Starts at 0, ends at stop-1
range(start, stop)      # Starts at start, ends at stop-1
range(start, stop, step) # Counts by step from start to stop-1
```


### 💡 Simple Example: The `range()` Function


````python
# Generate even numbers from 2 to 10 using range
# start=2, stop=11 (exclusive), step=2
for number in range(2, 11, 2):
    print("Even number:", number)
````

**Expected Output:**
```text
Even number: 2
Even number: 4
Even number: 6
Even number: 8
Even number: 10
```


### 🌍 Real-Life Example: The `range()` Function


````python
# Print a countdown timer for a product launch
# Starts at 5, goes down to 1, steps down by -1
for seconds in range(5, 0, -1):
    print("Launching in", seconds, "seconds...")

print("We are live! 🚀")
````

**Expected Output:**
```text
Launching in 5 seconds...
Launching in 4 seconds...
Launching in 3 seconds...
Launching in 2 seconds...
Launching in 1 seconds...
We are live! 🚀
```

## 🔹 Nested Loops

### Explanation
A nested loop is a loop inside another loop. The inner loop executes completely from start to finish for every single iteration of the outer loop.

### Syntax
```python
for outer_item in outer_seq:
    for inner_item in inner_seq:
        # Code block
```


### 💡 Simple Example: Nested Loops


````python
# Combine numbers and letters using nested loops
numbers = [1, 2]
letters = ['A', 'B']

# Outer loop
for num in numbers:
    # Inner loop
    for char in letters:
        print(num, char)
````

**Expected Output:**
```text
1 A
1 B
2 A
2 B
```


### 🌍 Real-Life Example: Nested Loops


````python
# Create a weekly training schedule for Saturday and Sunday
days = ["Saturday", "Sunday"]
exercises = ["Cardio", "Strength Training"]

# Loop through days (outer)
for day in days:
    print("--- Workout plan for", day, "---")
    # Loop through exercises (inner)
    for activity in exercises:
        print("Do:", activity)
````

**Expected Output:**
```text
--- Workout plan for Saturday ---
Do: Cardio
Do: Strength Training
--- Workout plan for Sunday ---
Do: Cardio
Do: Strength Training
```

## 🔹 The `break` Statement

### Explanation
The `break` statement immediately stops or exits a loop, skipping all remaining iterations. It is used to stop loop execution as soon as a target condition is met.

### Syntax
```python
for item in sequence:
    if condition:
        break
```


### 💡 Simple Example: The `break` Statement


````python
# Stop printing numbers when 3 is encountered
for num in [1, 2, 3, 4, 5]:
    if num == 3:
        print("Number 3 detected! Stopping loop.")
        break
    print("Current number:", num)
````

**Expected Output:**
```text
Current number: 1
Current number: 2
Number 3 detected! Stopping loop.
```


### 🌍 Real-Life Example: The `break` Statement


````python
# Search through inventory for a requested item and stop once found
warehouse_items = ["Laptop", "Monitor", "Keyboard", "Mouse"]
requested_item = "Keyboard"

for item in warehouse_items:
    print("Checking item:", item)
    if item == requested_item:
        print("Found the", requested_item, "! Stopping warehouse search.")
        break
````

**Expected Output:**
```text
Checking item: Laptop
Checking item: Monitor
Checking item: Keyboard
Found the Keyboard ! Stopping warehouse search.
```

## 🔹 The `continue` Statement

### Explanation
The `continue` statement skips the rest of the code inside the current loop iteration and immediately jumps to the next iteration. Unlike `break`, it does not stop the entire loop.

### Syntax
```python
for item in sequence:
    if condition:
        continue
```


### 💡 Simple Example: The `continue` Statement


````python
# Print numbers 1 to 5, skipping 3
for num in range(1, 6):
    if num == 3:
        # Skip this loop iteration
        continue
    print("Number:", num)
````

**Expected Output:**
```text
Number: 1
Number: 2
Number: 4
Number: 5
```


### 🌍 Real-Life Example: The `continue` Statement


````python
# Filter student attendance list and greet only present students
attendance = {"John": "Present", "Emma": "Absent", "Sophia": "Present"}

for student, status in attendance.items():
    if status == "Absent":
        # Skip absent students
        continue
    print("Welcome to class,", student, "!")
````

**Expected Output:**
```text
Welcome to class, John !
Welcome to class, Sophia !
```

## 🔹 The `pass` Statement

### Explanation
The `pass` statement is a null operation; it does absolutely nothing. It is used as a temporary placeholder when a loop or condition block is required syntactically but contains no code yet.

### Syntax
```python
for item in sequence:
    pass
```


### 💡 Simple Example: The `pass` Statement


````python
# Build a loop with a placeholder for a specific condition
for num in range(1, 4):
    if num == 2:
        # Code logic for number 2 will be added later
        pass
    print("Number:", num)
````

**Expected Output:**
```text
Number: 1
Number: 2
Number: 3
```


### 🌍 Real-Life Example: The `pass` Statement


````python
# Build a basic message receiver that ignores 'ping' messages
received_messages = ["hello", "ping", "buy_item"]

for msg in received_messages:
    if msg == "ping":
        # Pings require no action, so we pass
        pass
    else:
        print("Handling request:", msg)
````

**Expected Output:**
```text
Handling request: hello
Handling request: buy_item
```

## 🔹 Using `else` with Loops

### Explanation
Python loops can have an optional `else` block. The `else` block runs ONLY if the loop completes all of its iterations naturally. If the loop is exited early using a `break` statement, the `else` block is skipped.

### Syntax
```python
for item in sequence:
    # Code block
else:
    # Runs if loop completes without breaking
```


### 💡 Simple Example: Using `else` with Loops


````python
# Demonstrate loop else block
for num in range(1, 4):
    print("Iteration:", num)
else:
    # Runs because loop completed completely
    print("Loop finished completely with no breaks!")
````

**Expected Output:**
```text
Iteration: 1
Iteration: 2
Iteration: 3
Loop finished completely with no breaks!
```


### 🌍 Real-Life Example: Using `else` with Loops


````python
# Search security camera events for suspicious activity
camera_events = ["clean", "clean", "clean"]

for event in camera_events:
    if event == "intruder":
        print("Alert! Intruder spotted. Stopping scan.")
        break
else:
    # Runs only if no intruder was found in the loop
    print("Scan complete: No security threats detected.")
````

**Expected Output:**
```text
Scan complete: No security threats detected.
```

## 🔹 Infinite Loops (What they are & how to avoid them)

### Explanation
An infinite loop is a loop that runs forever because its condition never evaluates to `False`. To avoid infinite loops, always make sure the loop condition is updated inside the block so it eventually becomes `False`.

### Syntax
```python
while True:
    # This runs forever unless a break statement is executed
```


### 💡 Simple Example: Infinite Loops (What they are & how to avoid them)


````python
# Run a safe loop by updating the counter variable
counter = 1

while counter <= 3:
    print("Safe Loop count:", counter)
    # Increment counter to ensure loop condition (counter <= 3) becomes False
    counter += 1
````

**Expected Output:**
```text
Safe Loop count: 1
Safe Loop count: 2
Safe Loop count: 3
```


### 🌍 Real-Life Example: Infinite Loops (What they are & how to avoid them)


````python
# Simulating connection attempts to a server with a limit counter
connection_attempts = 1
max_allowed_attempts = 3

# If we forgot to increment connection_attempts, this would run forever!
while connection_attempts <= max_allowed_attempts:
    print("Attempting to connect to server... Attempt #" + str(connection_attempts))
    # Increment to prevent an infinite attempt loop
    connection_attempts += 1

print("Failed to connect after " + str(max_allowed_attempts) + " attempts.")
````

**Expected Output:**
```text
Attempting to connect to server... Attempt #1
Attempting to connect to server... Attempt #2
Attempting to connect to server... Attempt #3
Failed to connect after 3 attempts.
```

### 🛠️ Sum of First N Numbers


````python
# Calculate the sum of numbers from 1 to 5
limit_n = 5
running_sum = 0

for num in range(1, limit_n + 1):
    # Add the current number to the running total
    running_sum += num

print("Sum of numbers from 1 to 5 is:", running_sum)
````



### 🛠️ Multiplication Table


````python
# Print the multiplication table for number 5 up to 5
table_number = 5

for multiplier in range(1, 6):
    result = table_number * multiplier
    print(table_number, "x", multiplier, "=", result)
````

### 🛠️ Factorial of a Number


````python
# Calculate factorial of 5 (5! = 5 * 4 * 3 * 2 * 1)
target_number = 5
factorial_result = 1

for num in range(1, target_number + 1):
    factorial_result *= num

print("Factorial of 5 is:", factorial_result)
````

**Expected Output:**
```text
Factorial of 5 is: 120
```


### 🛠️ Reverse a String using a Loop


````python
# Reverse a string character-by-character using a loop
original_text = "Python"
reversed_text = ""

for char in original_text:
    # Add the character to the beginning of the new string
    reversed_text = char + reversed_text

print("Original:", original_text)
print("Reversed:", reversed_text)
````

**Expected Output:**
```text
Original: Python
Reversed: nohtyP
```


### 🛠️ Count Vowels in a String


````python
# Count how many vowels are in a given sentence
sentence = "Jupyter Notebook"
vowel_count = 0
vowels = "aeiouAEIOU"

for character in sentence:
    if character in vowels:
        vowel_count += 1

print("Sentence:", sentence)
print("Number of vowels:", vowel_count)
````

**Expected Output:**
```text

Number of vowels: 6
```


### 🛠️ Find the Largest Number in a List


````python
# Find the maximum value in a list without using max() function
score_list = [45, 82, 19, 98, 54]
largest_value = score_list[0]  # Initialize with the first item

for score in score_list:
    # Update largest_value if a higher value is found
    if score > largest_value:
        largest_value = score

print("Scores:", score_list)
print("Largest Score is:", largest_value)
````

**Expected Output:**
```text
Scores: [45, 82, 19, 98, 54]
Largest Score is: 98
```

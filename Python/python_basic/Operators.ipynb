# Operators in Python

## 1️⃣ What is Operators?
In programming, an **operator** is a special symbol that tells the computer to perform a specific mathematical, comparison, or logical action on values. The values or variables that the operator works on are called **operands**.

For example:
* In the expression `10 + 5`:
  * `+` is the **operator** (performing addition).
  * `10` and `5` are the **operands**.

---

## 2️⃣ Categories of Operators in Python
Python supports several types of operators, classified by the operations they perform:

### 🔑 Operators Quick Reference:

| Type | Operators | Purpose | Example |
| :--- | :--- | :--- | :--- |
| **Arithmetic** | `+`, `-`, `*`, `/`, `%`, `//`, `**` | Performs basic math calculations. | `5 ** 2` (25) |
| **Comparison** | `==`, `!=`, `>`, `<`, `>=`, `<=` | Compares two values; returns `True` or `False`. | `x > 10` |
| **Logical** | `and`, `or`, `not` | Combines conditional statements. | `x > 5 and y < 10` |
| **Assignment** | `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=` | Assigns values to variables (includes shorthand). | `x += 5` |
| **Identity** | `is`, `is not` | Checks if both variables reference the same memory object. | `a is b` |
| **Membership** | `in`, `not in` | Checks if a value is present in a sequence (list, tuple, str). | `3 in [1, 2, 3]` |
| **Bitwise** | `&`, `|`, `^`, `~`, `<<`, `>>` | Performs operations on binary representations of numbers. | `5 & 1` (1) |


```python
# Arithmetic Operators Demo
print("--- Arithmetic Operators ---")
a, b = 15, 4
print(f"Addition: {a} + {b} = {a + b}")
print(f"Division (Float): {a} / {b} = {a / b}")
print(f"Floor Division (Integer): {a} // {b} = {a // b}")
print(f"Modulus (Remainder): {a} % {b} = {a % b}")
print(f"Exponentiation (Power): {a} ** {b} = {a ** b}")

# Assignment & Membership Demo
print("\n--- Membership & Assignment ---")
numbers = [1, 2, 3, 4, 5]
print(f"Is 3 in the list? {3 in numbers}")
print(f"Is 10 not in the list? {10 not in numbers}")
```

## 3️⃣ Deep Dive: Floor Division (`//`) & Modulus (`%`) with Negatives

A common interview trick is asking how floor division and modulus operate on negative numbers.

### 💡 The Rule:
* **Floor Division (`//`)** rounds the result **down** to the next smallest integer (towards negative infinity, not towards zero).
* **Modulus (`%`)** in Python always shares the sign of the **denominator** (divisor). It is calculated using the formula:  
  `r = a - (b * (a // b))`

### 🔑 Examples:
* `5 // 2` is `2.5`, rounded down to `2`.
* `-5 // 2` is `-2.5`, rounded down to `-3`.
* `-5 % 2` is calculated as `-5 - (2 * (-5 // 2))` = `-5 - (2 * -3)` = `-5 + 6` = `1`.

```python
# Positive vs Negative Floor Division
print("Floor Division:")
print(f" 5 // 2  = {5 // 2}")
print(f"-5 // 2  = {-5 // 2}")   # Note: rounds down to -3

print("\nModulus:")
print(f" 5 % 2   = {5 % 2}")
print(f"-5 % 2   = {-5 % 2}")   # Output: 1 (shares sign of denominator 2)
print(f" 5 % -2  = {5 % -2}")  # Output: -1 (shares sign of denominator -2)
```

## 4️⃣ Logical Operators & Short-Circuit Evaluation

Logical operators (`and`, `or`, `not`) are used to combine conditional statements. In Python, logical operators are **lazy** and perform **short-circuit evaluation**.

### 💡 How Short-Circuiting Works:
* **`and` Operator**: If the first operand evaluates to `False`, Python immediately returns `False` without checking the second operand (because the expression can never be true).
* **`or` Operator**: If the first operand evaluates to `True`, Python immediately returns `True` without checking the second operand (because the expression is already guaranteed to be true).

### 🔑 Why is this asked in interviews?
It is used to prevent errors. For example, `x != 0 and (y / x) > 2` will not raise a `ZeroDivisionError` if `x` is `0`, because the first statement is `False` and the division is never evaluated.

```python
# Demonstrating Short-Circuit Evaluation

def check_second():
    print("Checking second operand...")
    return True

# 1. OR Short-Circuit: The first operand is True, so check_second() is NEVER called.
print("--- Test OR Short-Circuit ---")
result_or = True or check_second()
print(f"Result: {result_or}\n")

# 2. AND Short-Circuit: The first operand is False, so check_second() is NEVER called.
print("--- Test AND Short-Circuit ---")
result_and = False and check_second()
print(f"Result: {result_and}\n")

# 3. Safe division guard using short-circuiting
x = 0
# Without short-circuiting, evaluating (10 / x) would throw ZeroDivisionError.
# With short-circuiting, x != 0 is False, so the division is skipped!
is_safe = (x != 0) and (10 / x > 1)
print(f"Is division safe and evaluated? {is_safe}")
```

## 5️⃣ Operator Precedence, Associativity & Chaining

### ⚖️ Operator Precedence (Order of Operations):
When multiple operators appear in an expression, Python evaluates them in order of precedence (highest to lowest). Here is a simplified hierarchy:
1. Parentheses `()`
2. Exponentiation `**`
3. Unary positive/negative `+x`, `-x`
4. Multiplication, Division, Modulus, Floor Division `*`, `/`, `%`, `//`
5. Addition and Subtraction `+`, `-`
6. Comparisons, Identity, Membership `==`, `>`, `<`, `is`, `in`
7. Logical `not`, then `and`, then `or`

### 🔄 Operator Associativity:
If operators have the same precedence, they are evaluated based on their **associativity**:
* **Left-to-Right** (Most operators, like subtraction): `10 - 5 - 2` is evaluated as `(10 - 5) - 2` = `3`.
* **Right-to-Left** (Exponentiation `**`): `2 ** 3 ** 2` is evaluated as `2 ** (3 ** 2)` = `2 ** 9` = `512` (not `8 ** 2 = 64`).

### 🔗 Comparison Operator Chaining:
In Python, comparisons can be chained together.
For example, `1 < x < 10` is shorthand for `(1 < x) and (x < 10)`. It is evaluated cleanly without repeating `x`.

```python
# 1. Exponentiation right-to-left associativity
print(f"2 ** 3 ** 2 = {2 ** 3 ** 2}")  # 2 ** 9 = 512

# 2. Chaining Comparisons
x = 5
print(f"Is 1 < x < 10? {1 < x < 10}")  # True
```

## 6️⃣ Identity vs. Equality & Bitwise Operators

### ⚖️ Identity (`is`) vs. Equality (`==`):
* **`==`** compares the **values** of the operands.
* **`is`** compares the **memory locations (addresses)** of the operands.

### ⚡ Bitwise Operators:
Bitwise operators perform operations directly on binary levels of integers:
* `&` (Bitwise AND): Copies a bit to the result if it exists in both operands.
* `|` (Bitwise OR): Copies a bit if it exists in either operand.
* `^` (Bitwise XOR): Copies the bit if it is set in one operand but not both.
* `~` (Bitwise NOT): Flips the bits.
* `<<` (Left Shift): Shifts bits left, filling with zeros (multiplies by $2^n$).
* `>>` (Right Shift): Shifts bits right, discarding shifted bits (floor divides by $2^n$).

```python
# Bitwise Demonstration
# 5 in binary: 0101
# 3 in binary: 0011

x, y = 5, 3

print(f"x & y (AND): {x & y}")  # 0101 & 0011 = 0001 (1)
print(f"x | y (OR):  {x | y}")  # 0101 | 0011 = 0111 (7)
print(f"x ^ y (XOR): {x ^ y}")  # 0101 ^ 0011 = 0110 (6)

# Shifts
print(f"x << 1 (Left Shift): {x << 1}")   # 5 * 2 = 10 (1010)
print(f"x >> 1 (Right Shift): {x >> 1}")  # 5 // 2 = 2 (0010)
```

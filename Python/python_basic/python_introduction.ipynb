# Python Introduction & Core Rules

An interview-focused, concise guide to Python fundamentals: programming paradigm, syntax, semantics, and key syntax rules.

## 1️⃣ What is Python?

**Python** is a high-level, general-purpose, interpreted, and dynamically-typed programming language.

### 🔑 Core Characteristics for Interviews:
* **Interpreted**: Python code is executed line-by-line. Under the hood, Python source code (`.py`) is compiled into intermediate **bytecode** (`.pyc`), which is then executed by the **Python Virtual Machine (PVM)**.
* **Dynamically Typed**: Variable types are determined at runtime. You do not need to declare types explicitly before using them.
* **High-Level**: Python manages memory automatically using a **Garbage Collector** and abstracts away system hardware details, focusing on developer productivity and readability.

## 2️⃣ Syntax vs. Semantics

Understanding the distinction is crucial for explaining bugs and errors in interviews.

| Concept | Definition | Example Error |
| :--- | :--- | :--- |
| **Syntax** | The *grammar rules* that define structure. | `SyntaxError: invalid syntax` (e.g., missing `:` or unclosed parenthesis). |
| **Semantics** | The *meaning* or *logic* of the code. | Logical bug (e.g., using `+` instead of `*` for multiplication, or type mismatch). |

```python
# --- SYNTAX ERROR EXAMPLE ---
# Uncommenting the lines below causes a SyntaxError at compile/parse time:
# if True
#     print("Missing colon after True")

# --- SEMANTIC ERROR EXAMPLE ---
# The code compiles and runs perfectly (valid syntax), but the logic (semantics) is wrong:
def calculate_area(width, height):
    return width + height  # Semantic error: should be width * height

print("Calculated Area (with semantic bug):", calculate_area(5, 4))  # Output: 9 (Expected: 20)
```

## 3️⃣ Key Python Syntax Rules

### Rule A: Case Sensitivity
Python is strictly case-sensitive. Identifiers like `name`, `Name`, and `NAME` represent completely different objects/variables.

```python
name = "Amit"
Name = "Vijay"

print("name:", name)
print("Name:", Name)
print("Are they identical?", name == Name)  # Output: False
```

### Rule B: Indentation
Instead of curly braces `{}` (as in C/Java) or `begin`/`end` blocks, Python uses **indentation** to define the scope of control structures (loops, functions, classes).
* **Standard**: 4 spaces is the PEP 8 standard. Avoid tabs.
* **Consequences**: Mixing tabs and spaces will raise an `TabError` or `IndentationError`.

```python
def check_even(number):
    # 4 spaces indentation level for the function block
    if number % 2 == 0:
        # 8 spaces indentation level for the conditional block
        return "Even"
    else:
        return "Odd"

print("Is 4 even?", check_even(4))
```

### Rule C: Statements, Line Continuation, and Comments
* **Statement Termination**: Statements usually end with a newline. Semicolons `;` are optional and discouraged.
* **Line Continuation**: If a statement exceeds PEP 8's limit (79 characters), you can split it:
  * **Implicitly** inside parentheses `()`, brackets `[]`, or braces `{}` (Preferred).
  * **Explicitly** using a backslash `\` at the end of the line.
* **Comments & Documentation**:
  * **Single-line**: Prefixed with `#`.
  * **Docstrings**: Multi-line string literals (`"""..."""` or `'''...'''`) placed immediately after a function/class definition to document it (accessible via `__doc__`).

```python
# 1. Implicit line continuation (preferred style)
months = [
    "January", "February", "March",
    "April", "May", "June"
]

# 2. Explicit line continuation
long_sum = 10 + \
           20 + \
           30

# 3. Docstring example
def greet(user):
    """Takes a username and returns a welcome greeting."""
    return f"Hello, {user}!"

print(greet("Amit"))
print("Documentation:", greet.__doc__)
```

## 3️⃣ Deep Dive: List (`list`)

### 💡 Definition:
A **list** is an ordered, mutable, and dynamic sequence of elements. It allows duplicate values and can hold elements of different data types (heterogeneous).

### ⚙️ Internal Implementation (CPython):
* Under the hood, Python lists are implemented as **dynamic arrays** (arrays of pointers to other Python objects).
* **Over-Allocation**: To make appending elements fast, Python allocates extra memory space ahead of time. When the list fills up, Python copies elements to a new, larger block of memory.

### 🔑 Key Time Complexities:
* **Access (Indexing)**: $O(1)$
* **Append (add to end)**: Amortized $O(1)$ (constant time due to overallocation)
* **Insert/Delete (at index $i$)**: $O(n)$ (since all elements after it must be shifted)
* **Search / Membership check (`in`)**: $O(n)$

```python
# --- LIST COMPREHENSION & OPERATIONS ---
# A concise way to create lists
squares = [x**2 for x in range(5)]
print(f"Squares: {squares}")

# append() vs extend()
list_a = [1, 2]
list_b = [3, 4]

# append adds the entire object as a single element
list_a.append(list_b)
print(f"After append(): {list_a}")  # Output: [1, 2, [3, 4]]

# extend merges elements of the iterable
list_c = [1, 2]
list_c.extend(list_b)
print(f"After extend(): {list_c}")  # Output: [1, 2, 3, 4]
```

## 4️⃣ Deep Dive: Tuple (`tuple`)

### 💡 Definition:
A **tuple** is an ordered, immutable sequence of elements. It allows duplicates and heterogeneous data types. Once created, its items cannot be changed, added, or removed.

### ⚙️ Internal Implementation:
* Tuples are stored in a single block of memory with a fixed size (no over-allocation).
* **Performance Benefit**: This static nature makes tuples faster to create, consume less memory, and write-protects data compared to lists.

### 🔑 Key Time Complexities:
* **Access (Indexing)**: $O(1)$
* **Search / Membership check (`in`)**: $O(n)$

```python
from collections import namedtuple

# Named Tuple: self-documenting tuple with named fields
Point = namedtuple('Point', ['x', 'y'])
p = Point(10, 20)
print(f"Point coordinate: x={p.x}, y={p.y}")

# Tuple Unpacking
coordinates = (4, 5, 6)
x, y, z = coordinates
print(f"Unpacked values: x={x}, y={y}, z={z}")

# Modifying a tuple element containing a mutable object (frequent interview question)
try:
    bad_tuple = (1, 2, [3, 4])
    bad_tuple[2].append(5)  # The nested list can be modified!
    print(f"Modified nested list in tuple: {bad_tuple}")
except TypeError as e:
    print(f"Error: {e}")
```

## 5️⃣ Deep Dive: Set (`set`) & Frozenset

### 💡 Definition:
A **set** is an unordered collection of unique elements. It is mutable, but it can only contain **hashable** (immutable) elements. It does not support indexing or slicing.

### ⚙️ Internal Implementation:
* Python sets are implemented using **hash tables** (similar to dictionaries, but storing only keys without associated values).
* Lookups, additions, and deletions are incredibly fast due to hash function lookup.

### 🔑 Key Time Complexities:
* **Add / Remove**: $O(1)$ average
* **Search / Membership (`in`)**: $O(1)$ average (making it ideal for deduplication and lookups)

### ❄️ Frozenset:
A `frozenset` is an **immutable** version of a set. Since it is immutable, it is **hashable**, meaning you can use a `frozenset` as a key in a dictionary or as an element inside another set.

```python
# Set operations: Union, Intersection, Difference
set_x = {1, 2, 3}
set_y = {3, 4, 5}

print(f"Union (|): {set_x | set_y}")
print(f"Intersection (&): {set_x & set_y}")
print(f"Difference (-): {set_x - set_y}")

# Frozenset usage
f_set = frozenset([1, 2, 3])
# f_set.add(4) -> AttributeError (frozenset is immutable)
my_dict = {f_set: "Frozenset as Key"}
print(f"Dictionary with frozenset key: {my_dict}")
```

## 6️⃣ Deep Dive: Dictionary (`dict`)

### 💡 Definition:
A **dictionary** is an unordered (historically) or insertion-ordered (since Python 3.7) collection of key-value pairs. Keys must be unique and hashable.

### ⚙️ Internal Implementation:
* Implemented as a **hash table**.
* In Python 3.6+, a new compact dictionary layout is used:
  1. A small array of indices (`hash index array`).
  2. A dense array of key-value entries (`entries array`).
  This reduces memory overhead by 30-95% and preserves insertion order!

### 🔑 Key Time Complexities:
* **Lookup (by Key)**: $O(1)$ average
* **Insert / Delete**: $O(1)$ average

### 🛡️ Safe Key Retrieval:
* `dict.get(key, default)` returns `default` if the key is missing (instead of raising `KeyError`).
* `collections.defaultdict` provides a default value factory automatically for missing keys.

```python
from collections import defaultdict

# Dictionary Comprehension
square_dict = {x: x**2 for x in range(5)}
print(f"Square Dict: {square_dict}")

# defaultdict demo
grouped_words = defaultdict(list)
words = ["apple", "apricot", "banana", "blueberry"]

for word in words:
    first_char = word[0]
    grouped_words[first_char].append(word)

print(f"Grouped Words: {dict(grouped_words)}")
```

## 7️⃣ Essential Interview Questions on Data Structures

### Q1: What is the time complexity of searching in a List vs. Set?
* **List**: $O(n)$ time complexity because Python must scan the list element-by-element.
* **Set**: $O(1)$ average time complexity because Python uses a hash function to jump directly to the memory bucket.

---

### Q2: What is the difference between `sort()` and `sorted()`?
* **`list.sort()`**: In-place sorting. It modifies the original list and returns `None`. It is only available for lists.
* **`sorted(iterable)`**: Creates and returns a new sorted list from any iterable. The original collection remains unchanged.

---

### Q3: Why is a dictionary key required to be hashable?
Dictionaries use hash tables under the hood. To look up a key in $O(1)$ time, Python hashes the key to find its bucket. If the key was mutable (like a list), its contents could change, resulting in a different hash value, making it impossible to find the value in the hash table.

---

### Q4: Can a tuple containing a list be used as a dictionary key?
No. Even though a tuple is immutable, it is only hashable if **all** of its elements are hashable. Since a list is mutable (unhashable), a tuple like `(1, 2, [3, 4])` is unhashable and will raise a `TypeError`.

```python
# sort() vs sorted()
my_list = [3, 1, 4, 2]
sorted_new = sorted(my_list)
print(f"Original: {my_list}, Sorted Return: {sorted_new}")

my_list.sort()
print(f"After sort() in-place: {my_list}")

print("-" * 40)

# Unhashable tuple demo
try:
    invalid_key = (1, 2, [3, 4])
    bad_dict = {invalid_key: "value"}
except TypeError as e:
    print(f"Failed to use tuple as dict key: {e}")
```

## Panda

 ### What is Pandas?
Pandas is a Python library used for data analysis and manipulation, providing powerful data structures like Series and DataFrame to handle structured data efficiently.

👉 It is mainly used to:

- Read data (CSV, Excel, JSON, etc.)

- Clean data (remove errors, missing values)

- Analyze data (filter, group, calculate)

- Transform data into useful formats

### Why Pandas is Used?
Without Pandas, handling large datasets in Python is difficult.

With Pandas, you can:

- Work with thousands/millions of rows easily

- Perform operations in just one line of code

- Save a lot of time in data analysis

### Series
A Series in Pandas is a one-dimensional labeled array capable of holding data of any type, where each value is associated with an index.

👉 You can think of it like:

A single column of a table

Or a Python list with labels (index)

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {1:10, 2:20, 3:30}
```

```python
# Create series using list
pd.Series(my_list)
```

```python
# Create Series with label
pd.Series(my_list, index=labels)
```

```python
# Create Series using array
pd.Series(arr, index=labels)
```

```python
# Create Series using dictionary
pd.Series(d)
```

### Removing Missing Data
In real datasets, you often find missing values like:

NaN (Not a Number)

None

👉 Pandas provides functions to remove or handle these missing values easily.

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Vinod", "Rahul", "Amit", "Neha"],
    "Marks": [80, np.nan, 85, np.nan],
    "Age": [22, 25, np.nan, np.nan]
}

df = pd.DataFrame(data)
print(df)
```

#### How to Detect Missing Values

```python
df.isnull()      # True where value is missing
```

```python
df.isnull().sum()  # count missing values column-wise
```

```python
df.isnull().any()
```

```python
# Drop Rows with Missing Values
df.dropna()
```

```python
# Drop Columns with Missing Values
df.dropna(axis=1)
```

```python
# Drop Only Specific Columns
df.dropna(subset=["Marks"])
```

```python
# Drop Rows with All Missing Values
df.dropna(how="all")

# Removes rows only if all values are NaN
```

```python
# Use Threshold
# Keep rows with at least 2 non-null values
df.dropna(thresh=2)
```

```python
# Filling NaN
df.fillna(200)
```

```python
# fill different value on different column
values = {'Marks': 70, 'Age': 20}
df.fillna(value=values)
```

### Merging

Merging means combining two DataFrames based on a common column

```python
import pandas as pd

# DataFrame 1
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Vinod", "Rahul", "Amit"]
})

# DataFrame 2
df2 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Marks": [80, 90, 85]
})
pd.merge(df1, df2, on="ID")
```

#### Types of Merge

```python
# Inner Join (Default)
# Only matching data (common IDs)
pd.merge(df1, df2, on="ID", how="inner")
```

```python
# Left Join
# All data from left (df1)
pd.merge(df1, df2, on="ID", how="left")
```

```python
# Right Join
# All data from right (df2)
pd.merge(df1, df2, on="ID", how="right")
```

```python
# Outer Join
# All data from both (union)
pd.merge(df1, df2, on="ID", how="outer")
```

```python
# Merge on Different Column Names
# Used when column names are different
pd.merge(df1, df2, left_on="ID", right_on="ID")
```

### Concatenate

```python
import pandas as pd

df1 = pd.DataFrame({
    "Name": ["Vinod", "Rahul"],
    "Marks": [80, 90]
})

df2 = pd.DataFrame({
    "Name": ["Amit", "Neha"],
    "Marks": [85, 88]
})
```

```python
# Row-wise Concatenation
pd.concat([df1, df2])
```

```python
# Column-wise Concatenation
pd.concat([df1, df2], axis=1)
```

```python
# df.join()
import pandas as pd

df1 = pd.DataFrame({
    "Name": ["Vinod", "Rahul", "Amit"]
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    "Marks": [80, 90, 85]
}, index=[1, 2, 4])

result = df1.join(df2)
print(result)
```

### Join Types

```python
# Left Join (Default)
df1.join(df2, how="left")
```

```python
# Right Join
df1.join(df2, how="right")
```

```python
# Inner Join
df1.join(df2, how="inner")
```

```python
# Outer Join
df1.join(df2, how="outer")
```

### Difference between Concate and join

| Feature  | `concat()`                    | `join()`                  |
| -------- | ----------------------------- | ------------------------- |
| Purpose  | Attach DataFrames             | Combine using index       |
| Logic    | Stacking (no relation needed) | Matching (based on index) |
| Axis     | Row or Column                 | Mostly column-wise        |
| Use Case | Add rows/columns              | Combine related data      |


### GroupBy in Pandas
GroupBy in Pandas is used to split data into groups based on one or more columns and apply aggregation functions like sum, mean, or count.

```python
import pandas as pd

data = {
    "Name": ["Vinod", "Rahul", "Amit", "Neha", "Priya"],
    "Course": ["BCA", "MCA", "BCA", "MCA", "BCA"],
    "Marks": [80, 90, 85, 88, 92]
}

df = pd.DataFrame(data)
df
```

```python
# Basic GroupBy
df.groupby("Course")
```

```python
# GroupBy with Aggregation
# Average Marks per Course
df.groupby("Course")["Marks"].mean()
```

```python
# Total Marks per Course
df.groupby("Course")["Marks"].sum()
```

```python
# Count Students per Course
df.groupby("Course")["Name"].count()
```

```python
# Multiple Aggregations
df.groupby("Course")["Marks"].agg(["mean", "max", "min"])
```

```python
# Group by Multiple Columns
df.groupby(["Course", "Name"])["Marks"].sum()
```

```python
# Reset Index
df.groupby("Course")["Marks"].mean().reset_index()
```

### Aggregation in Pandas
Aggregation in Pandas is the process of applying functions like sum, mean, count, min, and max to summarize data.

```python
import pandas as pd

data = {
    "Name": ["Vinod", "Rahul", "Amit", "Neha"],
    "Marks": [80, 90, 85, 88],
    "Age": [22, 25, 23, 21]
}

df = pd.DataFrame(data)
```

```python
# Basic Aggregation
df["Marks"].sum()
df["Marks"].mean()
df["Marks"].max()
df["Marks"].min()
```

```python
# Aggregation on Multiple Columns
df[["Marks", "Age"]].mean()
```

```python
# Using agg() (Very Important 🔥)
df["Marks"].agg(["sum", "mean", "max"])
```

```python
# Aggregation with GroupBy
df.groupby("Name")["Marks"].sum()
```

```python
# Multiple Aggregations with GroupBy
df.groupby("Course")["Marks"].agg(["mean", "max", "min"])
```

### Pivot Table
A Pivot Table in Pandas is used to summarize data by grouping values and applying aggregation functions to create a structured table.

```python
import numpy as np 
import pandas as pd
```

```python
import pandas as pd

data = {
    "Name": ["Vinod", "Rahul", "Amit", "Neha", "Priya"],
    "Course": ["BCA", "MCA", "BCA", "MCA", "BCA"],
    "Marks": [80, 90, 85, 88, 92],
    "City": ["Delhi", "Mumbai", "Delhi", "Mumbai", "Delhi"]
}

df = pd.DataFrame(data)
df
```

```python
# Basic Pivot Table
pd.pivot_table(df, values="Marks", index="Course")
```

```python
# Multiple Columns
pd.pivot_table(df, values="Marks", index="Course", columns="City")
```

```python
# Using Different Aggregation Functions
pd.pivot_table(df, values="Marks", index="Course", aggfunc="sum")
```

```python
# Multiple Aggregations
pd.pivot_table(df, values="Marks", index="Course", aggfunc=["mean", "max"])
```

```python
# Fill Missing Values
pd.pivot_table(df, values="Marks", index="Course", columns="City", fill_value=0)
```

```python
# Count Example
pd.pivot_table(df, values="Name", index="Course", aggfunc="count")
```
